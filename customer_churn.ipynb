# Data Cleaning
From above we see that most of the columns are completely filled. We will drop columns from the data that are unnecessary for overall analysis.


```python
#We need to map a bunch of columns to numbers for easier analysis
#Male 0, Female 1
#Yes 1, No 0
#df = df.dropna()
#Remove non-necessary
ndf = df.drop(['CustomerID','Count','Country','State','City','Zip Code','Lat Long','Latitude','Longitude','Churn Label','Churn Reason'],axis=1)

#Mapping the data to be numeric
male_female_map = {'Male':0,'Female':1}
yes_no_map = {'No':0,'Yes':1}
multiple_lines_map = {'No':0, 'Yes':1, 'No phone service':2}
internet_service_map = {'No':0, 'DSL':1,'Fiber optic':2}
features_map = {'No':0, 'Yes':1, 'No internet service':2}
contract_map = {'Month-to-month':0,'Two year':1,'One year':2}
payment_method_map = {'Mailed check':0,'Electronic check':1,'Bank transfer (automatic)':2,'Credit card (automatic)':3}
#Remapping
ndf['Gender'] = ndf['Gender'].map(male_female_map)
yes_no = ['Senior Citizen','Partner','Dependents','Phone Service','Paperless Billing']
for i in yes_no:
    ndf[i] = ndf[i].map(yes_no_map)
ndf['Multiple Lines'] = ndf['Multiple Lines'].map(multiple_lines_map)
ndf['Internet Service'] = ndf['Internet Service'].map(internet_service_map)

features = ['Online Security','Online Backup','Device Protection','Tech Support','Streaming TV','Streaming Movies']
for i in features:
    ndf[i] = ndf[i].map(features_map)

ndf['Contract'] = ndf['Contract'].map(contract_map)
ndf['Payment Method'] = ndf['Payment Method'].map(payment_method_map)

#Convert all types to be float
#ndf = ndf.astype(float)

#Take a look at the dataframe
ndf
```

## Breakdown of Churn by Demographics

Let's try to visualize differences in customer churn by different demographics. To minimize confusion, I created my own function to make simple stacked bar plots in matplotlib. Each set of graphs explores some form of demographic that describes the customers and its relationship to churn.

```python
def make_stacked_bar(df, col_name,labela,labelb,labs):
    achurn = (df[col_name] == labela) & (df['Churn Label'] == 'Yes')
    bchurn = (df[col_name] == labelb) & (df['Churn Label'] == 'Yes')
    #male_churn
    acount = [i for i in achurn if i is True]
    anumchurn = len(acount)
    astay = df[col_name].value_counts()[labela] - anumchurn
    bcount = [i for i in bchurn if i is True]
    fchurn = len(bcount)
    fstay = df[col_name].value_counts()[labelb] - fchurn
    labels = labs
    ameans = [anumchurn, fchurn]
    bmeans = [astay, fstay]
    width = 0.35
    plt.bar(labels, ameans)
    plt.bar(labels, bmeans, bottom = ameans)
    plt.legend(['Churn','Stayed'])
    plt.title('Breakdown of Churn by ' + col_name)
    plt.show()


make_stacked_bar(df,'Gender','Male','Female',['Male','Female'])
make_stacked_bar(df,'Senior Citizen','Yes','No',['Senior Citizen','Non-Senior'])
make_stacked_bar(df,'Partner','Yes','No',['Has a Partner','No Partner'])
make_stacked_bar(df,'Dependents','Yes','No',['Has Dependents','No Dependents'])


```

# Confusion Matrix

## What is a confusion matrix?
It is a method of evaluating machine learning algorithms!

### Definitions

* True Positive: Predicted the customer would churn and they did churn.
* True Negative: Predicted the customer would not churn and they did not churn.
* Type I Error: Predicted the customer would churn and the customer did not churn.
* Type II Error: Predicted the customer would not churn but the customer did churn.

Each matrix square refers to one of these defintions.
True Positive: (1,1)
True Negative: (0,0)
Type I Error: (0,1)
Type II Error: (1,0)

#### Analysis
We use a metric called an F-Score to help evaluate a model.

$$ Recall = \frac{TruePositive}{TruePositive + TypeII} $$


$$ Precision = \frac{TruePositive}{TruePositive + TypeI} $$


$$ Fscore = \frac{2*Recall*Precision}{Recall + Precision} $$

```python
confusion = metrics.confusion_matrix(y_test, y_pred)
confusion
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(confusion,annot=True,cmap='viridis',linewidths=.5,fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

## F Score

An F Score is a way of evaluating a model objectively. The higher the F Score the more confident you can be in a model's accuracy.

```python
#Let's Calculate the F-Score

recall = confusion[1][1]/(confusion[1][1]+confusion[1][0])
precision = confusion[1][1]/(confusion[1][1]+confusion[0][1])
fscore = (2*recall*precision)/(recall+precision)
fscore
```
